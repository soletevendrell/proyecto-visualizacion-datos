## Limpieza para Contratos

```python
import pandas as pd
import glob

# Carpeta donde están tus CSVs (ajústala)
carpeta_csv = r"..\proyecto_paro\data\paro"
archivos = glob.glob(f"{carpeta_csv}\*.csv")

# for archivo in archivos:
#     print(archivo)
```

```python
# Lista para almacenar los DataFrames procesados
dfs_limpios = []

# Columnas numéricas a sumar
sum_cols = ['Total Contratos',
            'Contratos iniciales indefinidos hombres',
            'Contratos iniciales temporales hombres',
            'Contratos convertidos en indefinidos hombres',
            'Contratos iniciales indefinidos mujeres',
            'Contratos iniciales temporales mujeres',
            'Contratos convertidos en indefinidos mujeres',
            'Contratos Agricultura',
            'Contratos Industria',
            'Contratos Construcción',
            'Contratos Servicios']

# Columnas por las que agrupar
group_cols = ['Código de CA', 'Comunidad Autónoma', 'Codigo Provincia', 
              'Provincia', 'Codigo Municipio', 'Municipio', 'anio']
```

```python
for archivo in archivos:
    # Leer CSV
    df = pd.read_csv(archivo, sep=";", encoding='latin-1', skiprows=1, header=0)
        
    # Limpiar nombres de columnas
    df.columns = df.columns.str.strip()
    
    # Crear columna 'año' a partir de 'mes'
    df['anio'] = df['mes'].str.extract(r'(\d{4})').astype(int)
    
    # Resumen anual por municipio
    df_anual = df.groupby(group_cols)[sum_cols].sum().reset_index()
    
    # Añadir a la lista
    dfs_limpios.append(df_anual)

# Concatenar todos los años en un solo DataFrame
df_final = pd.concat(dfs_limpios, ignore_index=True)

# Mostrar resultado
df_final.head(5)
```

```python
display(df_final)
```

```python
# Guardar el DataFrame final en un CSV
df_final.to_csv(r"..\proyecto_paro\data\datos_compactos\contratos_anuales_por_municipio.csv", 
                index=False,  # No guardar el índice
                sep=";",      # Separador de columnas (igual que tus CSV originales)
                encoding='utf-8')  # Puedes usar 'latin-1' si lo abres en Excel
```

## Limpieza para paro

```python
import pandas as pd
import glob

# Carpeta donde están tus CSVs (ajústala)
carpeta_csv = r"..\proyecto_paro\data\paro"
archivos = glob.glob(f"{carpeta_csv}\*.csv")

# for archivo in archivos:
#     print(archivo)

archivo = archivos[0]  # Primer archivo

# Leer el CSV
df = pd.read_csv(archivo, sep=";", encoding='latin-1', skiprows=1, header=0)

# Limpiar espacios en nombres de columnas
df.columns = df.columns.str.strip()

# Mostrar nombres de columnas
print("Columnas del archivo:", archivo)
print(df.columns.tolist())
```

```python
# Lista para almacenar los DataFrames procesados
dfs_limpios = []

# Columnas numéricas a sumar

mean_cols = ['total Paro Registrado',
        'Paro hombre edad < 25',
        'Paro hombre edad 25 -45',
        'Paro hombre edad >=45',
        'Paro mujer edad < 25',
        'Paro mujer edad 25 -45',
        'Paro mujer edad >=45',
        'Paro Agricultura',
        'Paro Industria',
        'Paro Construcción',
        'Paro Servicios',
        'Paro Sin empleo Anterior']


# Columnas por las que agrupar
group_cols = ['Código de CA', 'Comunidad Autónoma', 'Codigo Provincia', 
              'Provincia', 'Codigo Municipio', 'Municipio', 'anio']
```

```python
for archivo in archivos:
    # Leer CSV
    df = pd.read_csv(archivo, sep=";", encoding='latin-1', skiprows=1, header=0)
    
    # Limpiar nombres de columnas
    df.columns = df.columns.str.strip()
    
    # Crear columna 'año' a partir de 'mes'
    df['anio'] = df['mes'].str.extract(r'(\d{4})').astype(int)

    for col in mean_cols:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    
    # Resumen anual por municipio
    df_anual = df.groupby(group_cols)[mean_cols].mean().reset_index().round(1)
    
    # Añadir a la lista
    dfs_limpios.append(df_anual)

# Concatenar todos los años en un solo DataFrame
df_final = pd.concat(dfs_limpios, ignore_index=True)

# Mostrar resultado
df_final.head(5)
# display(df_final)
```

```python
display(df_final)
df_final.to_csv(r"..\proyecto_paro\data\datos_compactos\paro_anual_por_municipio.csv", 
                index=False,  # No guardar el índice
                sep=";",      # Separador de columnas (igual que tus CSV originales)
                encoding='utf-8')  # Puedes usar 'latin-1' si lo abres en Excel
```
